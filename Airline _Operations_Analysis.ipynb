## Basic Analysis

### How many planes have more than 100 seats?

```python
pd.read_sql_query("""select aircraft_code, count (*) as num_seats
from seats group by aircraft_code Having num_seats >100""", connection)
```

### How did the number of tickets booked and the total amount earned changed over time?

```python
tickets = pd.read_sql_query("""select * from tickets inner join bookings
  on tickets .book_ref = bookings.book_ref""", connection)

tickets ['book_date'] = pd.to_datetime(tickets ['book_date'])
tickets ['date'] = tickets ['book_date'].dt.date
x = tickets.groupby('date')[['date']].count()
plt.figure(figsize = (18,6))
plt.plot(x.index, x['date'], marker = '^')
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Number of Tickets', fontsize = 20)
plt.grid('b')
plt.show()
```

```python
bookings = pd.read_sql_query("select * from bookings", connection)
bookings ['book_date'] = pd.to_datetime(bookings ['book_date'])
bookings ['date'] = bookings ['book_date'].dt.date
x = bookings.groupby ('date')[['total_amount']].sum()
plt.figure(figsize = (18,6))
plt.plot(x.index, x['total_amount'], marker = '^')
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Total Amount Earned', fontsize = 20)
plt.grid('b')
plt.show()
```

### Average Charges for Each Aircraft With Different Fare Conditions

## Occupancy Rate

### Total Revenue Per Year And The Average Revenue Per Ticket For Each Aircraft

```python
pd.read_sql_query(""" SELECT aircraft_code, ticket_count, total_revenue, total_revenue/ticket_count as avg_revenue_per_ticket from
(select aircraft_code, count(*) as ticket_count, sum(amount) as total_revenue
from ticket_flights join flights on ticket_flights.flight_id = flights.flight_id
group by aircraft_code) """, connection)
```

### Average Occupancy Per Aircraft

```python
occupancy_rate = pd.read_sql_query("""select a.aircraft_code, avg(a.seat_count) as booked_seats, b.num_seats, avg(a.seat_count)/b.num_seats as occupancy_rate from
(select aircraft_code, flights.flight_id, count(*) as seat_count from boarding_passes
inner join flights on boarding_passes.flight_id = flights.flight_id
group by aircraft_code, flights.flight_id) as a
inner join (select aircraft_code, count(*) as num_seats from seats
group by aircraft_code) as b on a.aircraft_code = b.aircraft_code
group by a.aircraft_code
""", connection)

occupancy_rate
```

### How Much Total Annual Turnover Could Increase by Giving All Aircrafts a 10% Higher Occupancy Rate

```python
occupancy_rate ['Inc occupancy rate'] = occupancy_rate ['occupancy_rate'] * 1.1
occupancy_rate
```
